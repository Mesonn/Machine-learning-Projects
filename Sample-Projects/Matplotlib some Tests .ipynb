```python
heart = pd.read_csv("heart-disease.csv")
heart.head()
```

```python
heart["age"].plot.hist(bins = 10)
```

```python
heart.plot.hist(figsize = (10, 30) , subplots = True);
```

```python
oldies = heart[heart["age"] > 50]
oldies.head()
```

```python
#pyplot method
oldies.plot(kind = 'scatter',
           x = 'age',
        y = 'chol',
           c = 'target')
```

```python
# OO Method 
fix, ax = plt.subplots(figsize = (10,6))
oldies.plot(kind = 'scatter',
                x = 'age',
                y = 'chol',
                c = 'target',
                ax = ax)
    
```

```python
## OO Method
fig, ax = plt.subplots(figsize = (10,6))

scatter = ax. scatter(x = oldies["age"],
                     y = oldies["chol"],
                     c = oldies["target"])
ax.set(title="Heart disease and cholesterol Levels",
        xlabel = "Age",
          ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(),title = "Target")
ax.axhline(oldies["chol"].mean(),linestyle = '--')
```

```python
oldies.head()
```

```python
plt.style.use('seaborn-whitegrid')


fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1,
                               figsize = (10,10),sharex= True)

scatter = ax0. scatter(x = oldies["age"],
                     y = oldies["chol"],
                     c = oldies["target"],
                      cmap = "winter")
ax0.set(title="Heart disease and cholesterol Levels",
       
          ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(),title = "Target")
ax0.axhline(oldies["chol"].mean(),linestyle = '--', c = 'red')
ax0.set_xlim([50,80])

scatter = ax1. scatter(x = oldies["age"],
                     y = oldies["thalach"],
                     c = oldies["target"],
                      cmap = "winter")
ax1.set(title="Heart disease and Max heart Rate",
        xlabel = "Age",
          ylabel="Max Heart rate")
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
ax1.legend(*scatter.legend_elements(),title = "Target")

ax1.axhline(oldies["thalach"].mean(),linestyle = '--',c = 'red')

fig.suptitle("Heart Disease Analysis",fontsize = 16, fontweight = "bold")
```
